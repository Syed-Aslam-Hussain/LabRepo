```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Connect to the SQLite database
conn = sqlite3.connect('survey-results-public.sqlite')
```

## Demo: Basic SQL Queries

```python
QUERY = "SELECT COUNT(*) FROM main"
df = pd.read_sql_query(QUERY, conn)
print(df)
```

```python
QUERY = """
SELECT name as Table_Name 
FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

#### Demo 3: Group Data by Age


```python
QUERY = """
SELECT Age, COUNT(*) as count 
FROM main 
GROUP BY Age 
ORDER BY Age
"""
df_age = pd.read_sql_query(QUERY, conn)
print(df_age)
```

## Visualizing Data


### Task 1: Visualizing the Distribution of Data


**1. Box Plot of `CompTotal` (Total Compensation)**


Use a box plot to analyze the distribution and outliers in total compensation.


```python
# your code goes here
Query="""SELECT CompTotal FROM main"""
df=pd.read_sql_query(Query,conn)
boxplt_data=df.dropna()
bx_data=boxplt_data.groupby('CompTotal').size()
bx_data.plot(kind='box',figsize=(10,6))
plt.title('Distribution and outliers in total compensation')
plt.xlabel('CompTotal')
plt.ylabel('Count')
plt.show() 
```

**2. Box Plot of Age (converted to numeric values)**


Convert the `Age` column into numerical values and visualize the distribution.


```python
# your code goes here
Query="""SELECT Age FROM main"""
df=pd.read_sql_query(Query,conn)
age_mapping ={
    'Under 18 years old':17,
    '18-24 years old':21,
    '25-34 years old':29,
    '35-44 years old':39,
    '45-54 years old':49,
    '55-64 years old':59,
    '65 years or older':70,
    'Prefer not to say':40
}
df['Age']=df['Age'].map(age_mapping)
df=df.dropna()
df = df.sort_values(by='Age', ascending=True)
plt.figure(figsize=(10,6))
plt.boxplot(df['Age'])
plt.title('Distribution in Age-group')
plt.ylabel('Age')
plt.show()
```

### Task 2: Visualizing Relationships in Data


**1. Box Plot of `CompTotal` Grouped by Age Groups:**


Visualize the distribution of compensation across different age groups.


```python
# your code goes here
Query="""SELECT CompTotal,Age FROM main"""
df=pd.read_sql_query(Query,conn)
box_data=df[['Age','CompTotal']].dropna()
data_gr=box_data.groupby('Age')['CompTotal'].size().reset_index()
plt.figure(figsize=(10,6))
sns.boxplot(x='Age',y='CompTotal',data=data_gr)
plt.title('Distribution of compensation across different age groups')
plt.xticks(rotation=45,ha='right')
plt.show()
```

**2. Box Plot of `CompTotal` Grouped by Job Satisfaction (`JobSatPoints_6`):**


Examine how compensation varies based on job satisfaction levels.


```python
# your code goes here
Query="""SELECT CompTotal, JobSatPoints_6 FROM main"""
df=pd.read_sql_query(Query,conn)
df=df.dropna()
df_gr=df.groupby('JobSatPoints_6')['CompTotal'].size().reset_index()
plt.figure(figsize=(14,8))
sns.boxplot(x='JobSatPoints_6',y='CompTotal',data=df_gr)
plt.title('Distribution of compenstions based on job satisfaction level',fontsize=20)
plt.xticks(rotation=90,ha='right')
plt.tight_layout()
plt.show()
```

### Task 3: Visualizing the Composition of Data


**1. Box Plot of `ConvertedCompYearly` for the Top 5 Developer Types:**


Analyze compensation across the top 5 developer roles.


```python
# your code goes here
Query="""SELECT ConvertedCompYearly,DevType FROM main ORDER BY ConvertedCompYearly DESC LIMIT 5"""
df=pd.read_sql_query(Query,conn)
plt.figure(figsize=(10,6))
sns.boxplot(x='DevType',y='ConvertedCompYearly',data=df)
plt.title('Distribution of compensation across the top 5 developer roles')
plt.xticks(rotation=45,ha='right')
plt.show()
```

**2. Box Plot of `CompTotal` for the Top 5 Countries:**


Analyze compensation across respondents from the top 5 countries.


```python
# your code goes here
Query="""SELECT CompTotal,Country FROM main GROUP BY Country ORDER BY CompTotal DESC LIMIT 5 """
df=pd.read_sql_query(Query,conn)
plt.figure(figsize=(10,6))
sns.boxplot(x='Country',y='CompTotal',data=df)
plt.title('Distribution of compensation across respondents from top 5 countries')
plt.xticks(rotation=45,ha='right')
plt.show()
```

### Task 4: Visualizing Comparison of Data


**1. Box Plot of CompTotal Across Employment Types:**


Analyze compensation for different employment types.


```python
# your code goes here
Query="""SELECT Employment,CompTotal FROM main GROUP BY Employment"""
df=pd.read_sql_query(Query,conn)
df=df.dropna()
df['Employment']=df['Employment'].str.split(';')
df= df.explode('Employment')
df = df.reset_index(drop=True)
plt.figure(figsize=(12,8))
sns.boxplot(x='Employment',y='CompTotal',data=df)
plt.title('Distribution of compensation for different employment types')
plt.xticks(rotation=45,ha='right')
plt.show()
```

**2. Box Plot of `YearsCodePro` by Job Satisfaction (`JobSatPoints_6`):**


Examine the distribution of professional coding years by job satisfaction levels.


```python
# your code goes here
Query="""SELECT YearsCodePro,JobSatPoints_6 FROM main GROUP BY YearsCodePro"""
df=pd.read_sql_query(Query,conn)
df=df.dropna()
plt.figure(figsize=(10,6))
sns.boxplot(x='YearsCodePro',y='JobSatPoints_6',data=df)
plt.title('Distribution of coding years by job satisfaction levels')
plt.xticks(rotation=45,ha='right')
plt.show()
```
