### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
df['EdLevel'].value_counts() # identifying inconsistent entries in EdLevel column
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
EdLevel_mapping={
'Bachelor’s degree (B.A., B.S., B.Eng., etc.)':'Bachelor',
'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)':'Master',
'Some college/university study without earning a degree' :'Student',
'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)':'Secondary',
'Professional degree (JD, MD, Ph.D, Ed.D, etc.)':'Professional',
'Associate degree (A.A., A.S., etc.)':'Associate',
'Primary/elementary school':'Elementary',
'Something else':'other'
}
# standardize entries in EdLevel column by mapping inconsistent values to consistent format
df['EdLevel']=df['EdLevel'].replace(EdLevel_mapping) 
df['EdLevel']
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
df_encoded_emp=pd.get_dummies(df['Employment']) # encodimg column using one-hot encoding
df_encoded_emp
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
df.isnull().sum().idxmax() # finding out the column with the highest number of missing values
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
import numpy as np
Mean=df['WorkExp'].mean()
df['WorkExp']=df['WorkExp'].replace(np.nan,Mean) # imputing missing values in the numerical column with the mean
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
# imputing missing values with the most frequent value
df['RemoteWork']=df['RemoteWork'].replace(np.nan,'Hybrid (some remote, some in-person)')
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
# applying Min-Max scaling to normalize the column
df['ConvertedCompYearly']=(df['ConvertedCompYearly']-df['ConvertedCompYearly'].min())/(df['ConvertedCompYearly'].max()-df['ConvertedCompYearly'].min())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
df['ConvertedCompYearly']=np.log1p(df['ConvertedCompYearly']) # reduced column skewness by using log-transform
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
# replace the string entries to numeric in YearsCodePro column
df['YearsCodePro']=df['YearsCodePro'].replace({'Less than 1 year':0.5,'More than 50 years':51})

df['YearsCodePro']= df['YearsCodePro'].apply(pd.to_numeric, errors='coerce') #Convert problematic columns to numeric

bins=np.linspace(np.min(df['YearsCodePro']),np.max(df['YearsCodePro']),5) # create bins with bins-width=5

label=['Entry-level','Mid-Level','Senior-Level','Lead'] # create a list, label based on experience level
# Creating a new column 'ExperienceLevel' based on the 'YearsCodePro' column
df['ExperienceLevel']=pd.cut(df['YearsCodePro'],bins,labels=label,include_lowest=True)

df['ExperienceLevel'].head(10)
```
