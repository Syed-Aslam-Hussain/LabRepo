### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
print(df.info())
df.describe()
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
df.isnull().sum()
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
plt.figure(figsize=(10,6))# set figur size value
sns.heatmap(df.isnull(),cbar=False, cmap='viridis')# visualize missing values using a heatmap
plt.title('Heatmap of Missing Values')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
 
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
print(f'Number of missing rows for columns LearnCodeOnline :{df['LearnCodeOnline'].isnull().sum()}')
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
import numpy as np
df['Employment'].value_counts().idxmax()#identify most frequency value in Employment column
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
#impute missing values in the Employment column with the most frequent value
df['Employment']=df['Employment'].replace(np.nan,'Employed, full-time')
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
distr_data=df.groupby('RemoteWork').size().reset_index(name='Count')# select RemoteWork columns for visualization
plt.figure(figsize=(12,6)) #set figure size
sns.barplot(data=distr_data,x='RemoteWork',y='Count',width=0.6)#visualize the distribution of RemoteWork
plt.title('Distribution of work-mode variation after imputation')#input chart title
plt.show()
```
