```python
# Load the data
df = pd.read_csv("survey-data.csv")

# Display the first few rows of the data to understand its structure
df.head()
```

### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
df=df.dropna(subset=['Age','SOPartFreq','JobSat'])
df['Scaled_JobSat']=df['JobSat']*15.0
plt.figure(figsize=(10,6))
plt.scatter(data=df,x='Age',y='SOPartFreq',s='Scaled_JobSat',alpha=0.3)
plt.title('Relationship between Age and frequency sized by job satisfaction level',fontsize=13)
plt.xticks(rotation=45,ha='right')
plt.xlabel('Respondents Age')
plt.ylabel('Frequency')
plt.grid(True,linestyle='--',alpha=0.5)
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here
df=df.dropna(subset=['ConvertedCompYearly','JobSat','Age'])
age_mapping ={
    'Under 18 years old':17,
    '18-24 years old':21,
    '25-34 years old':30,
    '35-44 years old':35,
    '45-54 years old':49,
    '55-64 years old':59,
    '65 years or older':70,
    'Prefer not to say':40
}
df['Age']=df['Age'].map(age_mapping)
df['Scaled_Age']=df['Age']*10.0
plt.figure(figsize=(10,6))
plt.scatter(data=df,x='JobSat',y='ConvertedCompYearly',s='Scaled_Age',alpha=0.3,c='coral')
plt.title('Relationship between compensation and job satisfaction sized by age')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Compensation')
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here
df=df.dropna(subset=['LanguageHaveWorkedWith','Age'])
df['LanguageHaveWorkedWith']=df['LanguageHaveWorkedWith'].str.split(';')
df=df.explode('LanguageHaveWorkedWith')
bubble_data=df.groupby(['Age', 'LanguageHaveWorkedWith']).size().reset_index(name = 'Frequency')
plt.figure(figsize=(16,16))
plt.scatter(data=bubble_data,x='LanguageHaveWorkedWith',y='Age',s='Frequency',alpha=0.4)
plt.title('Programming Language Preferences by Age (Size by Frequency)',fontsize=15)
plt.xlabel('Language have worked with')
plt.ylabel('Age')
plt.xticks(rotation=90,ha='right')
plt.show()               
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here
df=df.dropna(subset=['DatabaseWantToWorkWith','JobSat'])
df['DatabaseWantToWorkWith']=df['DatabaseWantToWorkWith'].str.split(';')
df=df.explode('DatabaseWantToWorkWith')
bub_data=df.groupby(['DatabaseWantToWorkWith','JobSat']).size().reset_index(name='Count')
plt.figure(figsize=(16,10))
plt.scatter(data=bub_data,x='DatabaseWantToWorkWith',y='JobSat',s='Count',c='darkgreen',alpha=0.5)
plt.title('Relationship between preferred databases want to work with and job satisfaction',fontsize=20)
plt.ylabel('Job Satisfaction Score',fontsize=15)
plt.xlabel('Database want to work with',fontsize=15)
plt.xticks(rotation=90,ha='right')
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here
df=df.dropna(subset=['ConvertedCompYearly','DevType','JobSat'])
plt.figure(figsize=(12,12))
sns.scatterplot(data=df,x='ConvertedCompYearly',y='DevType',size='JobSat', hue='JobSat', sizes=(10,300),alpha=0.6)
plt.title('Distribution of compensation across different roles sized by Job satisfaction',fontsize=15)
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here
df=df.dropna(subset=['NEWCollabToolsHaveWorkedWith','Age'])
df['NEWCollabToolsHaveWorkedWith'] =df['NEWCollabToolsHaveWorkedWith'].str.split(';')
df =df.explode('NEWCollabToolsHaveWorkedWith')
bub_df=df.groupby(['NEWCollabToolsHaveWorkedWith','Age']).size().reset_index(name='Frequency')
plt.figure(figsize=(10,6))
sns.scatterplot(data=bub_df,x='NEWCollabToolsHaveWorkedWith',y='Age',size='Frequency',hue='Frequency',sizes=(15,450),alpha=1)
plt.title('Relationship between the collaboration tools used and age groups',fontsize=15)
plt.xticks(rotation=45,ha='right')
plt.legend(title='Frequency', loc='upper left',bbox_to_anchor=(1, 1))
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here
df=df.dropna(subset=['WebframeWantToWorkWith','JobSat'])
df['WebframeWantToWorkWith']=df['WebframeWantToWorkWith'].str.split(';')
df=df.explode('WebframeWantToWorkWith')
df_bubble=df.groupby(['WebframeWantToWorkWith','JobSat']).size().reset_index(name='Count')
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_bubble,x='WebframeWantToWorkWith',y='JobSat',size='Count',hue='Count',sizes=(15,450),palette='viridis',alpha=0.7)
plt.title('Relationship between preferred web frameworks and job satisfaction(sized by frequency)')
plt.xticks(rotation=45,ha='right')
plt.legend(title='Count', loc='upper left',bbox_to_anchor=(1, 1))
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
##Write your code here
df=df.dropna(subset=['LanguageAdmired','Country'])
Top_countries=df['Country'].value_counts().index[:10]
df=df[df['Country'].isin(Top_countries)]
df['LanguageAdmired']=df['LanguageAdmired'].str.split(';')
df=df.explode('LanguageAdmired')
lang_data=df.groupby(['LanguageAdmired','Country']).size().reset_index(name='Frequency')
plt.figure(figsize=(15,14))
sns.scatterplot(data=lang_data,x='Country',y='LanguageAdmired',size='Frequency',sizes=(16,600),hue='Frequency',palette='Spectral',alpha=0.7)
plt.title('Distribution of admired technologies across different countries(sized by frequency)',fontsize=20)
plt.xticks(rotation=45,ha='right')
plt.legend(title='Frequency', loc='upper left',bbox_to_anchor=(1, 1))
plt.show()
```
